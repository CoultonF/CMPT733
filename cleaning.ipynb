```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('DataFiles/SGD-data.csv')
df.head()
```

```python
df1 = pd.melt(df, 
              id_vars=['Country Name', 'Series Name'], 
              value_vars=df.columns[4:], var_name='Year', 
              value_name='Indicator')
df1 = df1[df1['Indicator'].notnull()]
df1['Year'] = df1['Year'].str[:4]
df1
```

```python
df1.to_csv('cleaned_data.csv')
```
